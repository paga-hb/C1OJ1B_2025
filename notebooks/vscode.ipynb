## Visual Studio Code (VSCode)

Visual Studio Code (VSCode) är en texteditor (till skillnad från ett IDE som Visual Studio), utvecklad av Microsoft. VSCode använder extensioner (`Extensions`) för att utöka sin funktionalitet, vilket innebär att man kan utveckla program i valfritt programmeringsspråk, med valfritt ramverk, i VSCode.

Microsofts [Getting Started Tutorial](https://code.visualstudio.com/docs/getstarted/getting-started) är en utmärkt introduktion till Visual Studio Code, men vi utforskar Visual Studio Code tillsammans i denna notebooken.

Medan vi utforskar Visual Studio Code (VSCode), kommer vi att skapa en gRPC applikation <b>Todo</b> som består av en lösning (Solution) med följande projekt:

- `Todo.GrpcServer` som är ett ASP.NET Core gRPC Service projekt och består av server-sidan i gRPC applikationen.
- `Todo.GrpcClient` som är ett Console App projekt och består av klient-sidan i gRPC applikationen.
- `Todo.GrpcCommon` som ör ett Class Library projekt med gemensam kod för klienten och servern.

### Skapa `Todo` lösningen (Solution) i VSCode

För att skapa en lösning (Solution) i VSCode, måste man, för närvarande, använda `dotnet` CLI (Command Line Interface) verktyget.

Börja med att skapa en ny folder i ditt filsystem, flytta sedan in i foldern, och exekvera slutligen kommandot `dotnet new sln` i terminalen för att skapa en lösningsfil (Slution File) i den foldern:

```bash
dotnet new sln -n [name] -o [output]
```
här är `[name]` namnet på lösningen (Solution) och `[output]` är en sökväg till en folder i vilket lösningsfilen (Solution File) skall placeras. Om man vill skapa lösningsfilen (Solution File) i nuvarande folder, kan man använda:
```bash
dotnet new sln -n [name]
```

Slutligen anropar vi gRPC serverns metoder i `Program.cs`-

- Öppna file `Program.cs` under `Todo.GrpcClient` noden.
- Ersätt dess innehåll med nedanstående kod:

```csharp
using Grpc.Core;
using Grpc.Net.Client;
using System;
using System.Diagnostics;
using System.Net.Sockets;
using Todo.GrpcCommon;

namespace Todo.GrpcClient;

class Program
{
    static async Task  Main(string[] args)
    {
        var maxRetries = 10;
        var delay = TimeSpan.FromSeconds(1);
        var connected = false;

        using var channel = GrpcChannel.ForAddress("https://localhost:5001");
        var client = new Greeter.GreeterClient(channel);

        for (int i = 0; i < maxRetries; i++)
        {
            try
            {
                var reply = await client.SayHelloAsync(new HelloRequest { Name = "World!" });
                Console.WriteLine(reply.Message);
                connected = true;
                break;
            }
            catch (Exception ex) when (ex is HttpRequestException or SocketException or RpcException)
            {
                await Task.Delay(delay);
            }
        }

        if (!connected)
        {
            Console.WriteLine("Failed to connect to gRPC service after multiple attempts.");
            return;
        }

        Console.WriteLine("Press any key to exit...");
        Console.ReadKey();
    }
}
```

### Skapa en Bygg och Debug Profil så att vi kan debugga klienten och servern samtidigt

Låt oss först se hur man kan skapa en bygg profil i VSCode
- Höger-klicka på ett tomt område i `Explorer` (inte `Solution Explorer`) och välj `New Folder`.
  - Namnge foldern `.vscode` (med en inledande punkt `.`)
  - Detta är en speciell folder i VSCode som innehåller konfigurationsfiler för en given Workspace.
- Höger-klicka på `.vscode` foldern och välj `New File`
  - Namge filen `tasks.json` (denna filen innehåller bl.a. bygg profiler, t.ex. för att bygga (kompilera) våra projekt)
  - Öppna filen `tasks.json` och ersätt dess innehåll med nedanstående kod:

    ```csharp
    {
        "version": "2.0.0",
        "tasks": [
          {
            "label": "build Console App",
            "command": "dotnet",
            "type": "process",
            "args": [
              "build",
              "${workspaceFolder}/Todo.GrpcClient/Todo.GrpcClient.csproj"
            ],
            "problemMatcher": "$msCompile"
          },
          {
            "label": "build gRPC Service",
            "command": "dotnet",
            "type": "process",
            "args": [
              "build",
              "${workspaceFolder}/Todo.GrpcServer/Todo.GrpcServer.csproj"
            ],
            "problemMatcher": "$msCompile"
          },
          {
            "label": "build gRPC Common",
            "command": "dotnet",
            "type": "process",
            "args": [
              "build",
              "${workspaceFolder}/Todo.GrpcCommon/Todo.GrpcCommon.csproj"
            ],
            "problemMatcher": "$msCompile"
          },
          {
            "label": "build All",
            "dependsOn": ["build gRPC Common", "build gRPC Server", "build gRPC Client"],
            "dependsOrder": "sequence"
          }
        ]
    }
    ```
- Koden skapar en bygg-profil för respektive projekt (`build Console App`, `build gRPC Service`, `build gRPC Common`), samt för samtliga tre (`build All`).

Låt oss nu se hur man kan skapa en debug profil i VSCode
- Höger-klicka på `.vscode` foldern och välj `New File`
  - Namge filen `launch.json` (denna filen innehåller bl.a. debugg profiler, t.ex. för att debuga våra projekt)
  - Öppna filen `launch.json` och ersätt dess innehåll med nedanstående kod:

    ```csharp
    {
      "version": "0.2.0",
      "configurations": [
        {
          "name": "Console App",
          "type": "coreclr",
          "request": "launch",
          "preLaunchTask": "build Console App",
          // "preLaunchTask": "build All",
          "program": "${workspaceFolder}/Todo.GrpcClient/bin/Debug/net9.0/Todo.GrpcClient.dll",
          // "args": [],
          "cwd": "${workspaceFolder}/Todo.GrpcClient",
          // "console": "internalConsole",
          "console": "integratedTerminal",
          // "console": "externalTerminal",
          "stopAtEntry": false
        },
        {
          "name": "gRPC Service",
          "type": "coreclr",
          "request": "launch",
          "preLaunchTask": "build gRPC Service",
          // "preLaunchTask": "build All",
          "program": "${workspaceFolder}/Todo.GrpcServer/bin/Debug/net9.0/Todo.GrpcServer.dll",
          // "args": [],
          "cwd": "${workspaceFolder}/Todo.GrpcServer",
          // "console": "internalConsole",
          "console": "integratedTerminal",
          // "console": "externalTerminal",
          "stopAtEntry": false
        }
      ],
      "compounds": [
        {
          "name": "Run Console App + gRPC Service",
          "configurations": ["Console App", "gRPC Service"]
        }
      ]
    }
    ```
- Koden skapar en debugg-profil för klienten och servern (`Console App`, `bgRPC Service`), samt för båda (`Run Console App + gRPC Service`) i `compounds`.
  - Lägg märke till att `preLaunchTask": "build Console App"` i `Console App` anropar bygg-profilen `build Console App` innan Console App projektet debuggas.
  - Lägg märke till att `preLaunchTask": "build gRPC Service"` i `gRPC Service` anropar bygg-profilen `build gRPC Service` innan gRPC Service projektet debuggas.
  - Lägg märke till att respektive debugg-profil refererar till projektets kompilerade DLL via `program`.
  - Lägg märke till att respektive debugg-profil refererar till projektets huvudfolder via `cwd` (Current Working Directory).
  - Lägg märke till att respektive debugg-profil sätter `"console": "integratedTerminal"`vilket innebär att en ny inbyggd terminal skapas för varje projekt.
    - Det hade räckt att göra detta för våran Console App för att undvika undataget vi fick innan.
    - Om vi istället sätter `"console": "externalTerminal"`, öppnas en extern terminal istället (som i Visual Studio).
    - Vi kan också sätta `"console": "internalConsole"` för våran gRPC Service, eftersom den inte kräver en interaktiv terminal.
      - Denna inställningen innebär att utskrifter sker till `Debug Console` i VSCode.

Nu kan vi debugga båda projekten (klienten och servern) samtidigt (och undviker också undantaget vi fick innan).
- Byt till `Run and Debug` vyn huvudmenyn: `View -> Run`  (alternativt `Ctrl + Shift + D`, eller `Cmd + Shift + D` på en Mac).
  - I comboboxen (längst upp i `Run and Debug` vyn), välj `Run Console App + gRPC Service` (detta är våran `compound` debugg-profil).
  - Klicka därefter på den gröna "Run" knappen för att starta debugg-sessionen av gRPC klienten och gRPC servern.
- I den innbyggda `Terminalen`, lägg märke till utskriften från klienten:

```bash
Hello World!
Press any key to exit...
```
- Utskrifter från servern görs i en separat terminal.
- Avbryt debugg-sessionen via huvudmenyn: `Run -> Stop Debugging` (alternativt `Shift + F5`, eller genom att klicka på den röda *Stopp* knappen i `Debug Panelen`).
  - **Obs!** Du måste göra detta två gånger (en gång för att stoppa servern, och en gång för att stoppa klienten, om du inte terminerade klienten i terminalen via en knapptryckning).
